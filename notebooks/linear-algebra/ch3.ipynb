# 벡터 응용

## 상관관계와 코사인 유사도
- 상관계수: 두 변수 사이의 선형 관계를 정량화한 하나의 숫자
- 상관계수의 범위는 -1 ~ +1
- +1은 완벽한 양의 관계, -1은 완벽한 음의 관계, 0은 선형 관계가 없음
- 내적은 상관계수와 관련이 있고, 내적의 크기는 데이터 내의 수칫값의 크기와 관련이 있다.
- 상관계수가 -1 ~ +1 사이에 존재하려면 정규화가 필요하다.

### 각 변수의 평균중심화
- 각 데이터값에서 평균값을 빼는 것

### 벡터 노름 곱으로 내적을 나누기
- 분할(divisive) 정규화는 측정 단위를 제거하고 상관계수 최대 크기를 |1|로 조정한다.
- 피어슨 상관계수:
$$ p = \frac{\sum_{i=1}^{n}(x_i - \bar{x})(y_i - \bar{y})}{\sqrt{\sum_{i=1}^{n}(x_i - \bar{x})^2} \sqrt{\sum_{i=1}^{n}(y_i - \bar{y})^2}} $$

- 선형대수학으로 나타낸 피어슨 상관계수
- 물결은 정규화된 벡터를 의미(평균중심화)
1. 정규화된 벡터의 노름을 각각 구해서 곱한다.
2. 정규화된 벡터 X의 전치와 Y를 곱한다.
3. 정규화된 벡터의 노름의 곱으로 정규화된 벡터 X의 전치와 Y를 나눈다.
$$p = \frac{\tilde{X}^{\top}\tilde{Y}}
         {\|\tilde{X}\| \, \|\tilde{Y}\|}$$

- 코사인 유사도(cosine similarity): X와 Y의 노름을 곱하고 X와 Y의 내적을 나눈다.
$$ cos(\theta_{x,y}) = \frac{\alpha}{\|X\|\|Y\|}$$

- 피어슨 상관계수와 차이점은 코사인 유사도에는 첫 번째 정규화가 적용되어있지 않다.

## 시계열 필터링과 특징 탐지
- 필터링은 특징 탐지 기법이다. 커널(Kernel)이 시계열 신호의 일부와 일치하는 특징을 찾는다.
- 필터링된 결과는 또 다른 시계열이 되어 신호의 특성이 커널의 특성과 얼마나 일치하는지 알 수 있다.
- 커널과 시계열 신호 사이의 내적을 계산하는 것이 필터링 메커니즘
- 필터링할때 일반적으로 지역 특징 탐지를 해야하고 커널은 일반적으로 전체 시계열보다 훨씬 짧다.
- 커널과 동일한 길이의 짧은 데이터 조각과 커널 사이의 내적을 계산한다.
- 이 과정으로 필터링된 신호 구간에서 한 점이 생성되고 커널을 오른쪽으로 한 구간씩 이동시키면서 다른 신호 조각과 내적을 계산한다.
- 합성곱: 커널(필터)을 시계열(입력 신호) 상에서 지역적으로 이동(Sliding)시키면서, 각 위치에서 커널과 지역 데이터 조각(Local Patch) 사이의 내적을 반복적으로 계산하여 새로운 출력 신호를 생성하는 과정
