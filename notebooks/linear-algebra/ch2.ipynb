- 선형 종속적: 벡터 집합에서 하나의 벡터를 집합의 다른 벡터들의 선형 가중 결합으로 나타낼 수 있을때

## 부분 공간과 생성
- 유한한 벡터 집합 안에서 가중치로 선형 가중 결합할 수 있는 무한한 방법
- 생성: 가능한 모든 선형 가중 결합을 구성하는 메커니즘


```python
import matplotlib.pyplot as plt

vector = np.array([1, 3])

# 그래프 생성
fig, ax = plt.subplots(figsize=(10, 8))

# X축과 Y축을 0,0에서 교차하도록 명시적으로 그리기
ax.axhline(y=0, color='black', linewidth=1.5, alpha=0.8)  # X축
ax.axvline(x=0, color='black', linewidth=1.5, alpha=0.8)  # Y축

# 벡터 화살표 그리기 (원점에서 시작)
ax.quiver(0, 0, vector[0], vector[1], 
          angles='xy', scale_units='xy', scale=1, 
          color='blue', width=0.01, headwidth=4, headlength=5,
          label='벡터 [1,3]')

# 부분공간 그리기: 벡터 [1,3]이 생성하는 직선 (y = 3x)
# λ의 범위를 적절히 설정해서 직선 그리기
lambda_range = np.linspace(-2, 2, 100)
x_span = lambda_range * vector[0]
y_span = lambda_range * vector[1]

ax.plot(x_span, y_span, 'r--', alpha=0.7, linewidth=2, 
        label='부분공간: span{[1,3]} (y=3x)')

# 축 설정
ax.set_xlim(-0.5, 4)
ax.set_ylim(-0.5, 4)
ax.set_xlabel('X', fontsize=12)
ax.set_ylabel('Y', fontsize=12)
ax.set_title('벡터 [1,3]과 그 생성하는 부분공간', fontsize=14)
ax.grid(True, alpha=0.3)

# 축 비율 맞추기
ax.set_aspect('equal')

# 범례 추가
ax.legend()

# 벡터 값 표시
ax.text(vector[0]/2 + 0.1, vector[1]/2, f'[{vector[0]}, {vector[1]}]', 
        fontsize=12, ha='left', va='bottom', color='blue')

plt.show()
```

```python
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
import numpy as np

# 한글 폰트 설정
plt.rcParams['font.family'] = 'AppleGothic'
plt.rcParams['axes.unicode_minus'] = False

# 두 개의 3차원 벡터 정의
vector1 = np.array([1, 0, 2])
vector2 = np.array([-1, 1, 2])

# 3D 그래프 생성
fig = plt.figure(figsize=(12, 10))
ax = fig.add_subplot(111, projection='3d')

# 원점에 좌표축 그리기
ax.quiver(0, 0, 0, 4, 0, 0, color='black', alpha=0.3, linewidth=2, label='X축')  # X축
ax.quiver(0, 0, 0, 0, 4, 0, color='black', alpha=0.3, linewidth=2, label='Y축')  # Y축
ax.quiver(0, 0, 0, 0, 0, 4, color='black', alpha=0.3, linewidth=2, label='Z축')  # Z축

# 첫 번째 벡터 그리기 (파란색)
vector1_handle = ax.quiver(0, 0, 0, vector1[0], vector1[1], vector1[2],
                           color='blue', linewidth=3, arrow_length_ratio=0.1)

# 두 번째 벡터 그리기 (초록색)
vector2_handle = ax.quiver(0, 0, 0, vector2[0], vector2[1], vector2[2],
                           color='green', linewidth=3, arrow_length_ratio=0.1)

# 부분공간을 그리기 위해 그리드 생성
u = np.linspace(-2, 2, 10)
v = np.linspace(-2, 2, 10)
U, V = np.meshgrid(u, v)

# 평면 방정식: r = s*vector1 + t*vector2
X = U * vector1[0] + V * vector2[0]
Y = U * vector1[1] + V * vector2[1]
Z = U * vector1[2] + V * vector2[2]

plane_label = f"부분공간: span{{{vector1.tolist()}, {vector2.tolist()}}}"
plane_surface = ax.plot_surface(X, Y, Z, alpha=0.3, color='red')

# 축 설정
ax.set_xlim([-3, 5])
ax.set_ylim([-3, 5])
ax.set_zlim([-3, 5])

ax.set_xlabel('X', fontsize=12)
ax.set_ylabel('Y', fontsize=12)
ax.set_zlabel('Z', fontsize=12)
ax.set_title('3차원 벡터들과 그 생성하는 부분공간', fontsize=14)

ax.legend(
    handles=[vector1_handle, vector2_handle, plane_surface],
    labels=[
        f"벡터1 {vector1.tolist()}",
        f"벡터2 {vector2.tolist()}",
        plane_label,
    ],
    loc='upper left'
)

# 뷰포인트 설정 (더 잘 보이도록)
ax.view_init(elev=20, azim=45)

plt.tight_layout()
plt.show()
```

- 벡터 집합에서 생성되는 부분공간의 차원은 선형 독립 집합을 형성하는데 필요한 최소한의 벡터 수
- 벡터 집합이 선형 독립적이면 부분공간의 차원은 집합의 벡터 수와 동일
- 벡터 집합이 선형 종속적이면 벡터들로 생성된 부분 공간의 차원은 반드시 벡터 수보다 적다.
